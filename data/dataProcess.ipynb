```python
import pandas as pd 
import tensorflow as tf 
import tensorflow_datasets as tfds
```

```python
raw_data = pd.read_csv("./raw_dataset/IMDB Dataset.csv")
raw_data
```

```python
raw_data['sentiment'] = [1 if i == 'positive' else 0 for i in raw_data['sentiment']]
raw_data
```

```python
target_frame = pd.DataFrame(raw_data.iloc[:, -1]) 
target_frame
```

```python
input_frame = raw_data.iloc[:, 0:raw_data.shape[1]-1] 
input_frame
```

```python
tensor_data = tf.data.Dataset.from_tensor_slices((input_frame.values, target_frame.values)) 
sample = tensor_data.take(3) 
```

```python
tensor_data.save("./processed_data/")
```
